```python
import pickle
import numpy
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(processed)
```

```python
df.columns = ["x_shift", "y_shift", "distance", "lr", "repeat", "x_est", "y_est"]
```

```python
filter = df["distance"]==0
df.where(filter).groupby(["x_shift", "y_shift", "distance", "lr"]).mean()
```

```python
import numpy as np
df["err"] = np.power((df["x_shift"] + df["x_est"]),2) + np.power((df["y_shift"] + df["y_est"]),2)
```

```python
f1 = df["distance"]==0 
converted = df.where(f1).groupby(["x_shift", "y_shift", "distance", "lr"])["err"].agg("sum").reset_index()
```

```python
from scipy.ndimage.filters import maximum_filter1d as f1d
from scipy.ndimage.filters import gaussian_filter1d as f1gd
x_grouped = converted.groupby("lr")
plt.figure(figsize=(13,10))
for grp in x_grouped:
    srt = np.asarray(sorted(np.asarray([grp[1]["x_shift"], grp[1]["y_shift"], grp[1]["err"]]).transpose(), 
                 key=lambda x: np.sqrt(np.power(x[0],2) + np.power(x[1],2))))
    plt.plot(np.sqrt(np.power(srt[:, 0],2) + np.power(srt[:, 1],2)), f1d(f1gd(srt[:, 2], 10), 3), 
             label=grp[0], 
             color=[grp[0], 0, 0])
plt.yscale("log")
plt.legend(title="learning rate")
plt.ylabel("log(sum(|expected_shift - real_shift|)) over 20 runs")
plt.xlabel("norm distance -> sqrt(x^2 + y^2)")
plt.title("X,Y shift estimation error with regard to learning_rate")
plt.savefig("reports/01-lr2err.svg")
```
