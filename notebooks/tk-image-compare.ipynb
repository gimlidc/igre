```python
import matplotlib.pyplot as plt
import numpy as np

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
cnn = rgb2gray(plt.imread('CNN.png'))
dense = rgb2gray(plt.imread('NIR.png'))
origin = rgb2gray(plt.imread('origin.png'))
vis = plt.imread('visible.png')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(vis)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(origin, cmap='gray')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(cnn, cmap='gray')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(dense, cmap='gray')
```

```python
plt.style.use('classic')
diff = dense - cnn
plt.figure(figsize=(14,10))
plt.imshow(dense, cmap='gray')
plt.imshow(diff,
           cmap='seismic',
           vmin=-0.4,
           vmax=0.4,
           alpha=0.4*(np.abs(diff) > 0.06)
)
plt.show()
```

```python
plt.style.use('classic')
diff = cnn - dense
plt.figure(figsize=(14,10))
plt.imshow(cnn, cmap='gray')
plt.imshow(diff,
           cmap='seismic',
           vmin=-0.4,
           vmax=0.4,
           alpha=0.4*(np.abs(diff) > 0.06)
)
plt.show()
```
